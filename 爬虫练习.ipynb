```python
def get_page_content(url, params, headers):
    import requests
    import re
    import json
    
    response = requests.get(url, params=params, headers=headers)
    js_str = re.search('\{.*}',response.text)
    js_dict = json.loads(js_str.group())
    rateContent = [] #è¯„è®º
    auctionSku = [] #æ¬¾å¼
    rateDate = [] #æ—¶é—´
    dict_content = {}
    
    for i in range(len(js_dict['rateDetail']['rateList'])):
        rateContent.append([js_dict['rateDetail']['rateList'][i]['rateContent']])
        auctionSku.append([js_dict['rateDetail']['rateList'][i]['auctionSku']])
        rateDate.append([js_dict['rateDetail']['rateList'][i]['rateDate']])
    
    dict_content['rateContent'] = rateContent
    dict_content['auctionSku'] = auctionSku
    dict_content['rateDate'] = rateDate
    
    return dict_content
```
