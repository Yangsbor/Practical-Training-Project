```python
# jieba分词后保存在列表中
fileTrainSeg=[]
for i in range(len(fileTrainRead)):
    fileTrainSeg.append([' '.join(list(jieba.cut(fileTrainRead[i][2:-2],cut_all=False)))])
    if i % 100 == 0:
        print(i)
fileTrainSeg
```

```python
#去停用词前
fileSegWordDonePath = 'D:/fileSegWordDone.txt'
with open(fileSegWordDonePath,'w',encoding='utf-8') as fW:
    for i in range(len(fileTrainSeg)):
        fW.write(fileTrainSeg[i][0])
        fW.write('\n')

with open('D:/fileSegWordDone.txt','r',encoding='utf-8') as f:
    words = f.read().strip()
    print(len(words))
```

```python
#加载停用词
def load_stopwords(filepath):
    stopwords = []
    with open(filepath, 'r', encoding='utf-8') as fr:
        lines = fr.readlines()
        for line in lines:
            stopwords.append(line.strip())
    return stopwords
```

```python
#去停用词，保存分词结果到文件中
fileSegWordDonePath = 'D:/fileSegWordDone1.txt'
stopwords = load_stopwords('./stopwords/my_stopwords.txt')
with open(fileSegWordDonePath,'w',encoding='utf-8') as fw:
    for i in range(len(fileTrainSeg)):
        for word in fileTrainSeg[i][0].split(' '):
            if word not in stopwords:
                fw.write(word + ' ')
```

```python
#使用gensim.word2vec获取词向量
import warnings
import gensim
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence

#忽略警告
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
if __name__ == '__main__':
    input_cropus = 'D:/fileSegWordDone1.txt'
    out_vector = 'D:/cropusSegDone.vector'
    model = Word2Vec(LineSentence(input_cropus),size=50,window=5,
                     min_count=5,workers=4)
    model.wv.save_word2vec_format(out_vector,binary=False)
```

```python
#加载词向量

from gensim.models import KeyedVectors
word2vec_model_path = "D:/cropusSegDone.vector"
wv_from_text = KeyedVectors.load_word2vec_format(word2vec_model_path, binary=False)
word2vec_dict = {}
for word, vector in zip(wv_from_text.vocab, wv_from_text.vectors):
    if '.bin' not in word2vec_model_path:
        word2vec_dict[word] = vector
    else:
        word2vec_dict[word] = vector /np.linalg.norm(vector) 
for each in word2vec_dict:
    print(each,word2vec_dict[each])
```
