```python
import jieba
import pandas as pd
from snownlp import SnowNLP
from sklearn.model_selection import train_test_split

data = pd.read_csv('D:/comments.csv')

def cut_words(mytext):
    return " ".join(jieba.cut(mytext))
data['cut_comment'] = data['评论'].apply(cut_words)

def snow_result(comemnt):
    s = SnowNLP(comemnt)
    if s.sentiments >= 0.6:
        return 1
    else:
        return 0
    
data['snlp_result'] = data['评论'].apply(snow_result)
data
```

```python
def get_stopwords(stopwords_file_path):
    with open(stopwords_file_path,'r',encoding='utf-8') as f:
        stopwords = f.read()
    stopwords_list = stopwords.strip()
    stopwords_list = [i for i in stopwords_list]
    return stopwords_list

stopwords_file_path = './stopwords/my_stopwords.txt'
stopwords_list = get_stopwords(stopwords_file_path)
```

```python
X = data['cut_comment'] #分词后的评论
y = data['snlp_result'] #SnowNLP(text).sentiments

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)

from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(max_df = 0.8, 
                       min_df = 3, 
                       token_pattern=u'(?u)\\b[^\\d\\W]\\w+\\b', 
                       stop_words=frozenset(stopwords_list))

test = pd.DataFrame(vect.fit_transform(X_train).toarray(), columns=vect.get_feature_names())
test.head()
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

X_train_vect = vect.fit_transform(X_train)
nb.fit(X_train_vect, y_train)
train_score = nb.score(X_train_vect, y_train)
print(train_score)
```

```python
X_test_vect = vect.transform(X_test)
print(nb.score(X_test_vect, y_test))
```
